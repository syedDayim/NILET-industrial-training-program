```python
# X range 10 to 40, 30 
# Y range 10 to 80, 30
# x label = days
# y label = sales
# Make scatter graph, use symbol ^
x = np.random.randint(10, 41, 30)
y = np.random.randint(10, 80, 30)
plt.figure(figsize=(8,5))
plt.title("Assingment Graph", fontsize=16, color="blue")
plt.xlabel('Days', fontsize=10, color="green")
plt.ylabel('Sales', fontsize=10, color="green")
plt.scatter(x, y, color='red', label='assingment scatter graph', s=y, marker='^')
plt.show()
```

```python
# Pie chart
slices = [20, 80, 59, 99, 50]
names = ['Math', 'Sceince', 'Urdu', 'Sst', 'English']
plt.figure(figsize=(5,5))
plt.pie(slices, labels=names, autopct='%0.2f%%', explode=(0.01,0.01,0.01,0.2,0.01))
plt.show()
```
